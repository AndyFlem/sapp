```python
import numpy as np
import pandas as pd
import datetime
import math
from os import listdir
from os.path import isfile, join
import json
```

```python
out='../output/'
inb='../input/price/'
pref='dam'
inp = inb + pref + '/'
```

```python
dat=pd.DataFrame(columns=['date','hour','price','area'])
onlyfiles = [f for f in listdir(inp ) if isfile(join(inp, f))]
for f in onlyfiles:
    if (f.split('_')[0]==pref):
        #print(f)
        with open(join(inp, f), "r") as read_file:
            data = pd.DataFrame(json.load(read_file))
            dat=pd.concat([dat,data])

dat['price']=dat['price'].astype(float)
dat
```
